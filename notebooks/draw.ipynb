```python
import matplotlib.pyplot as plt

# Data points for noise level and corresponding performance
noise_levels = [5, 10, 15, 20, 25, 30, 50]
performance = [53.7, 52.8, 52.1, 50.4, 48.3, 46.2, 42.4]

# Baselines
full_model = 54.1
paraphrase = 53.7
qwen_plus_describer = 52.8

# Plotting
plt.figure(figsize=(7.2, 4.6))
plt.plot(noise_levels, performance, marker='o', label='Noisy semantics')

# Adding horizontal lines for baselines
plt.axhline(y=full_model, linestyle='-',color='tab:orange', label='Full (54.1)')
plt.axhline(y=paraphrase, linestyle='--',color='tab:green', label='Paraphrase (53.7)')
plt.axhline(y=qwen_plus_describer, linestyle='-.',color='tab:red', label='Qwen-Plus describer (52.8)')

plt.xlabel('Noise ratio (%)')
plt.ylabel('Avg F1 on 20 datasets')
plt.title('Robustness to Semantics Noise')
# plt.grid(True)
plt.legend(loc='best')

# Save as PDF

file_path = "/home/will/Projects/NERWork/Paper-LDG/notebooks/robustness_analysis_with_baselines.pdf"
# with open(file_path, "w", encoding="utf-8") as f:
#     pass
plt.savefig(file_path, format='pdf', bbox_inches='tight')
plt.close()

# file_path
```
